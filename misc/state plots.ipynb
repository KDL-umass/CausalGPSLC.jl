```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
data = pd.read_csv("../data/ISO/2018_smd_hourly.csv")
```

```python
summer_date = data.loc[103, "Date"]
summer_data = data[data["Date"] == summer_date]
summer_data = summer_data.append({"State":"CA", "RealTimeDemand":110000}, ignore_index=True)
summer_data = summer_data.append({"State":"OR", "RealTimeDemand":0}, ignore_index=True)
```

```python
summer_data
```

```python
winter_date = data.loc[0, "Date"]
winter_data = data[data["Date"] == winter_date]

winter_data = winter_data.append({"State":"CA", "RealTimeDemand":110000}, ignore_index=True)
winter_data = winter_data.append({"State":"OR", "RealTimeDemand":0}, ignore_index=True)
```

```python
fig = go.Figure(data=go.Choropleth(
    locations=winter_data['State'], # Spatial coordinates
    z = winter_data['RealTimeDemand'], # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = 'Greens',
    showscale=False
))

lon_min = 286.2
lon_max = lon_min + 7

fig.update_layout(
    geo = dict(
        resolution=50,
        showframe=False,
        showcoastlines=False,
        landcolor="white",
        lonaxis_range=[lon_min, lon_max],
        lataxis_range=[ 40.9, 47.6],
        domain = dict(x=[0, 1], y=[0, 1])
    )
)
fig.write_image("../figures/winter_demand.png")
```
